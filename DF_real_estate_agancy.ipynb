```python
import math

import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path ='/home/oksana/Repository/DF_real_estate_agancy/_data.csv'
df= pd.read_csv(path)
```

```python
# Первые пять строчек данных
df.head(10)
```

```python
# Данные  представлены 23368 строками с 24 признаками
print(df.shape)
```

```python
# Наименования столбцов (признаков) следующие:
print(df.columns.tolist())
```

```python
# Числовые значения (тип float, int64) только у двух признаков: ID  объявления, Высота потолков, м.
# Остальные 22 признака имеют тип object, поэтому требуют значительной предобработки.
print(df.info())
```

```python
# Ниже приведена статистика count,mean, std, min, 25%-50%-75% percentile, max только для этих двух признаков.
print(df.describe())
```

```python
# Количество уникальных значений для каждого столбца
df.nunique()
```

```python
# Подсчет пропусков по колонкам
for col in df.columns:
    print('Кол-во пропусков в колонке %s = %d' % (col, df[col].isna().sum()))
```

```python
# Подсчет уникальных значений по колонкам с выводом наименований колонок
for column in df.columns:
    unique_count = df[column].nunique()
    print(f"Колонка '{column}' имеет {unique_count} уникальных значений.")
```

```python
# 35,5% квартир - однокомнатные, 37,9% - двухкомнатные, 19% - трехкомнатныеб, остальные 4,5,6 комнатные
print(df['Количество комнат'].value_counts())
df['Количество комнат'].count()
```

```python
sns.barplot(y='Количество комнат', x='Unnamed: 0', color = 'purple', data=df.groupby('Количество комнат').count());
```
